### 구글 뉴스 데이터

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
url = "https://news.google.com/topics/CAAqJggKIiBDQkFTRWdvSUwyMHZNRGx1YlY4U0FtdHZHZ0pMVWlnQVAB?hl=ko&gl=KR&ceid=KR%3Ako"
page = urlopen(url)

soup = BeautifulSoup(page, "html.parser")
```

## 뉴스

```python
articles = soup.find_all("a", class_="DY5T1d")
len(articles)
```

```python
num = []
gubun = []
data = []
cnt = 0
for items in articles:
    cnt = cnt + 1
    print(items.text)
    num.append(cnt)
    gubun.append("세계")
    data.append(items.text)
```

```python
dict_dat = {"no":num, "구분":gubun, "title":data}
import pandas as pd 
dat_xls = pd.DataFrame(dict_dat)
dat_xls
```

```python
dat_xls.to_excel("data_xls.xlsx", sheet_name='Sheet_name_1', index=False)
```
