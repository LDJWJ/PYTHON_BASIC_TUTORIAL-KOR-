```python
import seaborn as sns
sns.set()
tips = sns.load_dataset("tips")
tips
```

## 팁과 식사 금액의 관계를 살펴보자

```python
### relplot () 함수는 많은 다른 통계적 관계를 시각화하기 위해 명명 되었기 때문에 
### 이름이 붙여졌습니다. 산점도가이를 수행하는 데 매우 효과적인 방법이지만 
### 한 변수가 시간의 척도를 나타내는 관계는 선으로 표시하는 것이 좋습니다.
```

```python
sns.relplot(x="total_bill", 
            y="tip", 
            data=tips)
```

```python
sns.relplot(x="total_bill", 
            y="tip",
            hue="smoker",
            data=tips)
```

```python
sns.relplot(x="total_bill", 
            y="tip", 
            col="time",
            hue="smoker", style="smoker", size="size",
            data=tips);
```

## 요일별 식사 총 금액의 분포는 어떨까?

### catplot
* 범주형 변수를 시각화하기 위한 plot()
* 하나의 숫자 변수(수치형 값)과 범주형 변수(구분되어진 값)관계 표현

```python
sns.catplot(x="day", 
            y="total_bill", 
            hue="smoker", 
            data=tips);
```

```python
sns.catplot(x="day", 
            y="total_bill", 
            hue="sex", 
            data=tips);
```

```python
sns.catplot(x="day", 
            y="total_bill", 
            hue="sex", 
            kind="bar",
            data=tips);
```

## 두 개의 그래프를 통해 비교 분석을 해 보자.


```python
import matplotlib.pyplot as plt
f, axes = plt.subplots(1, 2, sharey=True, figsize=(6, 4))
sns.boxplot(x="day", y="tip", data=tips, ax=axes[0])
sns.scatterplot(x="total_bill", y="tip", hue="day", data=tips, ax=axes[1]);
```

```python
## 세계의 그래프 
```

```python
import matplotlib.pyplot as plt
f, axes = plt.subplots(2, 2, sharey=True, figsize=(12, 10))
sns.boxplot(x="day", y="tip", data=tips, ax=axes[0][0])
sns.scatterplot(x="total_bill", y="tip", hue="day", data=tips, ax=axes[0][1]);
sns.barplot(x="day", y="total_bill", data=tips , ax=axes[1][0])
sns.barplot(x="day", y="total_bill", hue="sex", data=tips , ax=axes[1][1])
```

```python
sns.barplot(x="time", y="tip", data=tips,
              order=["Dinner", "Lunch"])
```

### 식사 인원별 전체 금액은 어떨까?

```python
ax = sns.barplot("size", y="total_bill", data=tips, palette="Blues_d")
```

### barplot의 주말과 주중의 식사 금액은 어떨지 구분해 보자.

```python
tips['weekend'] = tips['day'].isin(['Sat', 'Sun'])
ax = sns.barplot(x="day", y="total_bill", hue="weekend",
                    data=tips, dodge=False)
```

### 대표적인 데이터 셋 : iris
* pairplot()를 사용하여 변수간의 관계를 확인해 보자.

```python
iris = sns.load_dataset("iris")
iris
```

### 꽃잎(petal)과 꽃받침(sepal)의 길이의 데이터 분포를 확인해 보자.

```python
sns.jointplot(x="sepal_length", y="petal_length",  data=iris);  
```

```python

sns.pairplot(data=iris, hue="species");
```

## 여객 데이터를 이용한 시각화 

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

# Load the example flights dataset and conver to long-form
flights_long = sns.load_dataset("flights")
flights = flights_long.pivot("month", "year", "passengers")

# Draw a heatmap with the numeric values in each cell
f, ax = plt.subplots(figsize=(9, 6))
sns.heatmap(flights, annot=True, fmt="d", linewidths=.5, ax=ax)
```

### 타이타닉 데이터 셋 

```python
titanic = sns.load_dataset("titanic") 
```

### 각 카테고리 별로 데이터가 얼마나 있는지 확인해 보기

```python
sns.countplot(x="class", data=titanic)
plt.title("class of titanic")
plt.show()
```
